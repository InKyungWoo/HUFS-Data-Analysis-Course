```python
# 1. 캐글 데이터 찾기

import pandas as pd

# 파일 읽어오기
df = pd.read_csv('iris.data.csv', header=None)
df.head()    # 상위 5행 
```

```python
# 열 이름 변경
df.columns = ['Sepal_length','Sepal_width','Petal_length','Petal_width','Species'] 
df.head()
```

```python
# 2-1. 데이터프레임에서 필요한 데이터 찾기

df['Species'][0]
```

```python
# 2-2. Species 열 제외하고 출력(1)

df_drop_Species = df.drop(labels='Species',axis=1)
df_drop_Species
```

```python
# 2-3. Species 열 제외하고 출력(2)

df_drop_Species2 = df.loc[:,df.columns!='Species']
df_drop_Species2
```

```python
# 2-4. 열 추출하기

df_selected = df.loc[:,['Sepal_length','Species']]
df_selected
```

```python
# 2-5. 행 추출하기

df_selected2 = df_selected.loc[10:13,]
df_selected2
```

```python
# 3-1. 문자열 데이터를 찾아서 바꾸기

data = df['Species'][0].split("-")
data[1]
```

```python
# 3-2. 문자열 제거

data2 = df['Species'][0].replace('Iris-','')
data2
```

```python
# 3-3. 제거한 문자열 새로운 column으로 추가하기 : replace 함수 사용

df['Species_1'] = df['Species'].str.replace('Iris-','')
df
```

```python
# 3-4. 제거한 문자열 새로운 column으로 추가하기 : split 함수 사용

df['Species_2'] = df['Species'].str.split("-").str[1]
df
```

```python
# 4-1. 데이터의 도수 계산하기 : for문 사용

dic = {}

for _ in df['Species_2']:
    if _ in dic:
        dic[_] += 1
    else:
        dic[_] = 1

print(dic)
```

```python
# 4-2. 데이터의 도수 계산하기 : collections 모듈의 Counter 함수를 사용

from collections import Counter
Counter(df['Species_2'])
```

```python
# 4-3. 데이터의 도수 계산하기 : pd Series의 value_counts 함수를 사용

df['Species_2'].value_counts()
```

```python
# 4-4. 막대그래프로 나타내기

species_cnt = df['Species_2'].value_counts()
species_cnt.plot(kind='bar',title='Frequency of Species', rot=0, fontsize=15, width=0.2, color='violet')
```

```python
# 4-5. WordCloud로 시각화하기(1)

string_data = ''
for data in df['Species_2']:
    string_data =  string_data + ' ' + data
    
# WordCloud 생성
from wordcloud import WordCloud, STOPWORDS 
wordcloud = WordCloud(width = 400, height = 400, 
                     ).generate(string_data) 
  
# WordCloud 출력
import matplotlib.pyplot as plt

plt.figure(figsize = (4, 4), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout() 
plt.show()
```

```python
# 4-6. WordCloud로 시각화하기(2)

dict_data = Counter(df['Species_2'])

# WordCloud 생성
from wordcloud import WordCloud, STOPWORDS 
wordcloud = WordCloud(width = 400, height = 400, 
                      background_color ='black', 
                      stopwords = set(STOPWORDS), 
                      max_words=3,
                      max_font_size = 100,
                     ).generate_from_frequencies(dict_data) 
  
# WordCloud 출력                    
plt.figure(figsize = (4, 4), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout() 
plt.show()
```
