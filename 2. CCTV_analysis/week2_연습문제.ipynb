```python
# 연습문제3. 월별 데이터를 집계하기

# 3-1. ‘ENDTIME’ 의 MONTH 추출하여 열 추가

df['MONTH'] = df['ENDTIME'].str.split('-').str[1]
df
```

```python
# 3-2. 1월 데이터의 ‘INCOUNT’ 데이터의 평균값을 구하기

df[df['MONTH']=="01"]['INCOUNT'].mean()
```

```python
# 3-3. 12월(이전달) 과 1월의 ‘OUTCOUNT’ 데이터의 평균값

df1 = df.loc[(df['MONTH']=="01") | (df['MONTH']=="12")]
df1.groupby('MONTH')['OUTCOUNT'].mean()
```

```python
# 3-4. 12월과 1월 ‘INCOUNT’ 와 ‘OUTCOUNT’ 데이터의 min, MAX 구하기

outcnt = df1.groupby('MONTH')['OUTCOUNT'].agg(['min','max'])
incnt = df1.groupby('MONTH')['INCOUNT'].agg(['min','max'])

pd.concat([incnt, outcnt], axis=1, keys=['INCOUNT','OUTCOUNT'])
```

```python
# 4-5. 12월과 1월의 ‘OUTCOUNT’ 데이터의 평균값을 seaborn 라이브러리의 barplot 함수를 사용하여 시각화

from matplotlib import pyplot as plt
import seaborn as sns

outcnt_avg = df1.groupby('MONTH')['OUTCOUNT'].mean()

sns.barplot(x=outcnt_avg.index, y=outcnt_avg.values)

# 근데 1월은 너무 작아서 안보임..
```
